```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
mobile = tf.keras.applications.mobilenet.MobileNet()   #importing deep learning model - pretrained  mobilenet
```

Pre-processing of the image

```python
filename = 'desktop//India_tiger_count.jpg'
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(filename, target_size = (224,224))
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

# pre-processing

```python
resized_img = image.img_to_array(img)
final_img = np.expand_dims(resized_img, axis = 0) #adding fourth dimension
final_img = tf.keras.applications.mobilenet.preprocess_input(final_img)
```

```python
final_img.shape
```

```python
predictions = mobile.predict(final_img)
```

```python
print(predictions)
```

```python
from tensorflow.keras.applications import imagenet_utils #providing labels
```

```python
results = imagenet_utils.decode_predictions(predictions) #decoding
```

```python
print(results)
```
